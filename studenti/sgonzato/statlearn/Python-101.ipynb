## 2. Variabili e Tipi di Dati

Le **variabili** sono come dei contenitori usati per memorizzare dati. Ogni variabile ha un nome e un valore.
Python ha diversi **tipi di dati** incorporati, i più comuni sono:
- **Stringhe (`str`):** Sequenze di caratteri, come testo. Si scrivono tra virgolette singole (`'`) o doppie (`"`). Esempio: `"ciao"`, `'mondo'`.
- **Numeri Interi (`int`):** Numeri interi, positivi o negativi, senza decimali. Esempio: `10`, `-3`.
- **Numeri con Virgola Mobile (`float`):** Numeri con la parte decimale. Esempio: `3.14`, `-0.5`.
- **Booleani (`bool`):** Rappresentano uno dei due valori: `True` (vero) o `False` (falso).

### Extra: Casting di tipi di dato

**Il casting** permette di convertire una variabile da un tipo a un altro. Ad esempio, puoi trasformare un `int` in `float` o una `stringa` in un numero se compatibile.

Esempi comuni:
- Da intero a float: `float(5)`
- Da float a intero (scarta i decimali): `int(3.75)`
- Da stringa a intero: `int('23')` (solo se la stringa contiene cifre!)
- Da numero a stringa: `str(42)`

### List Comprehension

La **list comprehension** è un modo conciso per creare liste in Python. La sintassi è:
```python
[espressione for elemento in sequenza if condizione]
```

Esempio pratico:
- Creare una lista dei quadrati dei numeri da 0 a 4:
```python
quadrati = [x**2 for x in range(5)]
# Risultato: [0, 1, 4, 9, 16]
```

**Esercizio 1:** Crea una lista chiamata `doppi` che contenga i doppi dei numeri da 1 a 5 utilizzando la list comprehension.
Stampa la lista ottenuta.


````python
# Scrivi qui il tuo codice
doppi = [x*2 for x in range(1,6)]
print(doppi)
````

**Riscontro atteso:** Dovresti vedere [2, 4, 6, 8, 10]



**Esercizio 2:** Utilizza una list comprehension per creare una lista `solo_pari` che contenga solo i numeri pari da 0 a 10 inclusi. Stampa la lista.


````python
# Scrivi qui il tuo codice
solo_pari = [x for x in range(11) if x % 2 == 0]
print(solo_pari)
````

**Riscontro atteso:** Dovresti vedere [0, 2, 4, 6, 8, 10]

- Un commento su una singola riga inizia con `#`.
- Un commento su più righe (o una docstring per funzioni/classi) può essere racchiuso tra triple virgolette `""" ... """` o `''' ... '''`.
